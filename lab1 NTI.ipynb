```python
url= "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
df_newdata = pd.read_csv(url)
df_newdata
```

```python
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']
```

```python
raw_dataset = pd.read_csv(url, names=column_names,
                      na_values = "?", comment='\t',
                      sep=" ", skipinitialspace=True)
# skip for intial space ay line bad2 b space cancelo
# null here is ? not fady
# seperator here is space
```

```python
raw_data2 = raw_dataset.dropna()
# if we drope we can't get it again
```

```python
raw_data2
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mns= MinMaxScaler()
raw_data3=mns.fit_transform(raw_data2)
```

```python
raw_data3 = pd.DataFrame(raw_data3, columns=raw_data2.columns)
```

```python
x=raw_data3.drop('MPG',axis=1)
y=raw_data3['MPG']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

#MLP

```python
import keras
from sklearn.neural_network import MLPRegressor
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, input_dim=7, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer='sgd', metrics=['mse'])
```

```python
traine_model = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2 ,verbose=1)
```

```python
import matplotlib.pyplot as plt
plt.plot(traine_model.history['loss'])
plt.plot(traine_model.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
score = model.evaluate(x_test, y_test)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
