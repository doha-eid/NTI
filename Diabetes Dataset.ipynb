# MLP Deep Learning


```python
import keras
from keras.models import Sequential # Changed 'keras.model' to 'keras.models'
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, input_dim=10, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['binary_accuracy'])
```

```python
train_mode_db = model.fit(x_train, y_train, epochs=100, batch_size=16, validation_split=0.2, shuffle=True)
```

```python
score = model.evaluate(x_test, y_test)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
import matplotlib.pyplot as plt
plt.plot(train_mode_db.history['loss'])
plt.plot(train_mode_db.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

#PCA TO Feature reduction

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Assuming 'x' and 'y' are your original features and target variable

# Initialize and fit the PCA model
PCA_model = PCA(n_components=5)
PCA_model.fit(x)

# Transform the data to principal components
principal_components = PCA_model.transform(x)

# ... (previous code) ...

# Create a DataFrame with the principal components and target variable
df_pca = pd.DataFrame(principal_components, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5']) # Include column names for all 5 principal components
df_pca['target'] = y

# Split the data into training and testing sets
x_train_pca, x_test_pca, y_train_pca, y_test_pca = train_test_split(
    df_pca[['PC1', 'PC2', 'PC3', 'PC4', 'PC5']], df_pca['target'], test_size=0.2, random_state=42 # Include all principal components in the training and testing data
)

# ... (rest of the code) ...

# Initialize and fit the Logistic Regression model
pca_log_reg = LogisticRegression()
pca_log_reg.fit(x_train_pca, y_train_pca)

# Make predictions on the test set
y_pred_pca = pca_log_reg.predict(x_test_pca)

# Evaluate the model
print(classification_report(y_test_pca, y_pred_pca))
```
